#### We want to know cities and states with Temperature above 15 degress Celcius

```python
cities = []
states= []
def filtrar(lista, dato):
    for i in range(len(temp['Temperature'])):
        if temp['Temperature'][i] > 15:
            lista.append(temp[dato][i])
    return lista

print(filtrar(cities, 'City'))
print(filtrar(states, 'State'))
```

#### Find the mean and the standard deviation of the temperature of each state.

*Hint: Use functions from Data Manipulation lesson*

```python
def estados(estado):
    lista = []
    for i in range(len(temp['State'])):
        if temp['State'][i] == estado:
            lista.append(temp['Temperature'][i])
    return lista

tem_ny = estados('New York')
tem_con = estados('Connecticut')
tem_nj = estados('New Jersey')

ny_media = np.mean(tem_ny)
con_media = np.mean(tem_con)
nj_media = np.mean(tem_nj)
print('MEDIAS')
print(ny_media, con_media, nj_media)
ny_std = np.std(tem_ny)
con_std = np.std(tem_con)
nj_std = np.std(tem_nj)
print('DESVIACIONES ESTANDAR')
print(ny_std, con_std, nj_std)
```

#### Find the minimum, mean and the maximum of all numeric columns for each Department.

##### Hint: Use functions from Data Manipulation lesson

```python
sal_it = []
sal_hr = []
sal_sales = []
años_it = []
años_hr = []
años_sales = []

for i in range(len(employee['Department'])):
    if employee['Department'][i] == 'IT':
        sal_it.append(employee['Salary'][i])
        años_it.append(employee['Years'][i])
    if employee['Department'][i] == 'HR':
        sal_hr.append(employee['Salary'][i])
        años_hr.append(employee['Years'][i])
    if employee['Department'][i] == 'Sales':
        sal_sales.append(employee['Salary'][i])
        años_sales.append(employee['Years'][i])
        
print('DEPARTAMENTO IT: ')
print('Salario: ')
print('Minimo: ' + str(min(sal_it)), 'Máximo: ' + str(max(sal_it)), 'Media: ' + str(np.mean(sal_it)))
print('Años: ')
print('Minimo: ' + str(min(años_it)), 'Máximo: ' + str(max(años_it)), 'Media: ' + str(np.mean(años_it)))
print('\n')
print('DEPARTAMENTO HR: ')
print('Salario: ')
print('Minimo: ' + str(min(sal_hr)), 'Máximo: ' + str(max(sal_hr)), 'Media: ' + str(np.mean(sal_hr)))
print('Años: ')
print('Minimo: ' + str(min(años_hr)), 'Máximo: ' + str(max(años_hr)), 'Media: ' + str(np.mean(años_hr)))
print('\n')
print('DEPARTAMENTO Sales: ')
print('Salario: ')
print('Minimo: ' + str(min(sal_sales)), 'Máximo: ' + str(max(sal_sales)), 'Media: ' + str(np.mean(sal_sales)))
print('Años: ')
print('Minimo: ' + str(min(años_sales)), 'Máximo: ' + str(max(años_sales)), 'Media: ' + str(np.mean(años_sales)))
```
